```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Category": ["A", "A", "B", "B", "C"],
    "Sales": [100, 200, 300, 150, 400],
    "Profit": [20, 50, 80, 30, 100]
}
df = pd.DataFrame(data)

#Sales analysis
sales_summary = df.groupby("Category")["Sales"].sum()
print("Sales by Category")
print(sales_summary)

# Profit analysis
profit_summary = df.groupby("Category")["Profit"].sum()
print("\nProfit by Category:")
print(profit_summary)

# Visualization1
sales_summary.plot(kind="bar")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()

#Visulaization2
profit_summary.plot(kind="bar")
plt.title("Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```
